```python
import tensorflow as tf
import matplotlib.pyplot as plt
from tensorflow.keras import Model
from tensorflow.keras.layers import Conv2D, Dense, MaxPooling2D, Dropout, Flatten,GlobalAveragePooling2D
from tensorflow.keras.models import Sequential
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.callbacks import ReduceLROnPlateau
from tensorflow.keras.layers import Input, Lambda, Dense, Flatten
from tensorflow.keras.models import Model
from tensorflow.keras.applications.inception_v3 import InceptionV3
from tensorflow.keras.applications.inception_v3 import preprocess_input
from tensorflow.keras.preprocessing import image
from tensorflow.keras.preprocessing.image import ImageDataGenerator,load_img
from tensorflow.keras.models import Sequential
import numpy as np
from glob import glob
```

```python
train_datagen = ImageDataGenerator(rescale=1./255,
                                  shear_range=0.2,
                                  zoom_range=0.2,
                                  horizontal_flip=True)
train_datagen = ImageDataGenerator(rescale=1./255)
```

```python
#Training & Testing Spliting
training_set = train_datagen.flow_from_directory(r'dataset/train',
                                                target_size=(224, 224),
                                                batch_size=32,
                                                class_mode='categorical')
```

```python
test_set = train_datagen.flow_from_directory(r'dataset/test',
                                                target_size=(224, 224),
                                                batch_size=32,
                                                class_mode='categorical')
```

```python
inception_v3 = tf.keras.applications.InceptionV3(weights='imagenet', include_top=False, input_shape=(224, 224, 3))
```

```python
for layer in inception_v3.layers[: -15]:
    layer.trainable = False
```

```python
x = inception_v3.output
x = Flatten()(x)

x = Dense(units=512, activation='relu')(x)
x = Dropout(0.3)(x)
x = Dense(units=512, activation='relu')(x)
x = Dropout(0.3)(x)

output = Dense(units=5, activation='softmax')(x)
model = Model(inception_v3.input, output)

model.summary()
```

```python
loss = tf.keras.losses.CategoricalCrossentropy()
optimizer = tf.keras.optimizers.Adam(learning_rate=0.001)
model.compile(optimizer=optimizer, loss=loss, metrics=['accuracy'])
```

```python
histry = model.fit(training_set,
                            validation_data=test_set,
                            epochs=50,
                            steps_per_epoch=len(training_set),
                            validation_steps=len(test_set),
                            )
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.plot(histry.history['accuracy'], label='Training Accuracy')
plt.plot(histry.history['val_accuracy'], label='Validation Accuracy')
plt.title('Model Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(histry.history['loss'], label='Training Loss')
plt.plot(histry.history['val_loss'], label='Validation Loss')
plt.title('Model Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
from flask import Flask, request, jsonify, render_template
from tensorflow.keras.models import load_model
import numpy as np
from PIL import Image
import io
import os
import cv2
from io import BytesIO


app = Flask(__name__)

model = load_model('my_model.h5')

@app.route('/')
def index():
    return render_template('index.html')

@app.route('/predict', methods=['POST'])
def predict():
    if 'file' not in request.files:
        return jsonify({'error': 'No file part'})
    
    file = request.files['file']
    
    if file.filename == '':
        return jsonify({'error': 'No selected file'})
    
    img = Image.open(BytesIO(file.read()))
    
    img = img.resize((224, 224))

    img_array = np.array(img)

    if len(img_array.shape) == 2:  # Grayscale image
        img_array = cv2.cvtColor(img_array, cv2.COLOR_GRAY2BGR)

    img_array = img_array / 255.0  # Normalize to range [0, 1]

    img_array = np.expand_dims(img_array, axis=0)

    # Make prediction
    predictions = model.predict(img_array)

    classes = ['Normal', 'Stage 1', 'Stage 2', 'Stage 3','Stage 4']
    predicted_class_index = np.argmax(predictions, axis=1)[0]
    predicted_class_name = classes[predicted_class_index]
    print(predicted_class_name)
    return jsonify({'prediction': predicted_class_name})

if __name__ == '__main__':
    app.run(debug=False)
```

```python
from flask import Flask, request, jsonify, render_template, send_file
from tensorflow.keras.models import load_model
import numpy as np
import tensorflow as tf
import cv2
import matplotlib.pyplot as plt
from PIL import Image
import io
import os

app = Flask(__name__)

# Load the trained model
model = load_model('my_model.h5')

# Function to preprocess the image
def preprocess_image(image):
    img = image.resize((224, 224))  # Resize for InceptionV3
    img_array = np.array(img)
    if len(img_array.shape) == 2:  # Convert grayscale to RGB
        img_array = cv2.cvtColor(img_array, cv2.COLOR_GRAY2BGR)
    img_array = np.expand_dims(img_array / 255.0, axis=0)  # Normalize
    return np.array(img), img_array

# Grad-CAM function to generate heatmap
import cv2
import numpy as np
import tensorflow as tf
import os

def generate_gradcam_heatmap(img_array, original_img, model, layer_name="conv2d_93"):
    # Ensure original image is in the correct format
    if len(original_img.shape) == 2:  # If grayscale, convert to 3-channel BGR
        original_img = cv2.cvtColor(original_img, cv2.COLOR_GRAY2BGR)

    # Ensure original image has 3 channels
    if original_img.shape[-1] == 1:  # If single channel, convert to 3-channel
        original_img = np.stack([original_img] * 3, axis=-1)

    # Get last convolutional layer
    grad_model = tf.keras.models.Model([model.input], [model.get_layer(layer_name).output, model.output])

    # Compute gradients
    with tf.GradientTape() as tape:
        conv_outputs, predictions = grad_model(img_array)
        class_index = np.argmax(predictions[0])  # Get predicted class
        loss = predictions[:, class_index]  # Focus on predicted class

    grads = tape.gradient(loss, conv_outputs)  # Compute gradients
    pooled_grads = tf.reduce_mean(grads, axis=(0, 1, 2))  # Global average pooling

    # Compute activation map
    activation_map = np.dot(conv_outputs[0], pooled_grads.numpy())

    # Normalize activation map
    activation_map = np.maximum(activation_map, 0)  # ReLU: remove negative values
    activation_map /= np.max(activation_map)  # Normalize to [0, 1]

    # Resize heatmap to match original image size
    heatmap = cv2.resize(activation_map, (original_img.shape[1], original_img.shape[0]))

    # Apply Gaussian Blur to smooth heatmap and avoid edge artifacts
    heatmap = cv2.GaussianBlur(heatmap, (25, 25), 0)  # Adjust kernel size if needed

    # Convert heatmap to color
    heatmap = np.uint8(255 * heatmap)  # Scale between 0 and 255
    heatmap = cv2.applyColorMap(heatmap, cv2.COLORMAP_JET)  # Apply colormap

    # **Segment Knee Joint Region (Masking Unwanted Areas)**
    mask = np.zeros_like(heatmap, dtype=np.uint8)
    h, w, _ = mask.shape
    center_x, center_y = w // 2, h // 2  # Find center
    radius = min(h, w) // 3  # Define radius for knee region

    cv2.circle(mask, (center_x, center_y), radius, (255, 255, 255), -1)  # Circular mask

    # Apply the mask to the heatmap
    heatmap = cv2.bitwise_and(heatmap, mask)

    # Ensure original image and heatmap have the same dimensions
    if original_img.shape[:2] != heatmap.shape[:2]:  # Resize if needed
        heatmap = cv2.resize(heatmap, (original_img.shape[1], original_img.shape[0]))

    # Convert grayscale images to color
    if len(original_img.shape) == 2 or original_img.shape[-1] == 1:
        original_img = cv2.cvtColor(original_img, cv2.COLOR_GRAY2BGR)

    # Overlay heatmap on original image
    overlayed_img = cv2.addWeighted(original_img, 0.6, heatmap, 0.5, 0)

    # Save the heatmap image
    heatmap_path = os.path.join("static", "gradcam_heatmap.jpg")  # Save in static folder
    cv2.imwrite(heatmap_path, overlayed_img)

    return heatmap_path, class_index

@app.route('/')
def index():
    return render_template('index.html')

@app.route('/predict', methods=['POST'])
def predict():
    if 'file' not in request.files:
        return jsonify({'success': False, 'error': 'No file uploaded'})

    file = request.files['file']

    if file.filename == '':
        return jsonify({'success': False, 'error': 'No selected file'})

    try:
        img = Image.open(io.BytesIO(file.read()))
        original_img, img_array = preprocess_image(img)

        heatmap_path, predicted_class_index = generate_gradcam_heatmap(img_array, original_img, model)

        classes = ['Normal', 'Stage 1', 'Stage 2', 'Stage 3', 'Stage 4']
        predicted_class_name = classes[predicted_class_index]

        return jsonify({'success': True, 'prediction': predicted_class_name, 'heatmap_url': f'/{heatmap_path}'})

    except Exception as e:
        return jsonify({'success': False, 'error': str(e)})


@app.route('/heatmap')
def heatmap():
    return send_file("static/heatmap.jpg", mimetype='image/jpeg')

if __name__ == '__main__':
    app.run(port=5500,debug=False)
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def kmeans_segmentation(image_path, k=3):
    image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    image = cv2.resize(image, (256, 256))

    # Flatten image and apply K-means
    pixels = image.reshape(-1, 1)
    pixels = np.float32(pixels)
    
    criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 10, 1.0)
    _, labels, centers = cv2.kmeans(pixels, k, None, criteria, 10, cv2.KMEANS_RANDOM_CENTERS)
    
    segmented_image = centers[labels.flatten()].reshape(image.shape)
    
    return segmented_image.astype(np.uint8)

# Test function
seg_image = kmeans_segmentation("test9.png", k=3)
plt.imshow(seg_image, cmap='gray')
plt.title("K-Means Segmentation")
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.models import load_model
import tensorflow.keras.backend as K

# Load trained classification model
classifier_model = load_model("my_model.h5")

def segment_knee_watershed(image_path):
    # Read image
    image = cv2.imread(image_path)
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Apply thresholding
    _, thresh = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

    # Noise removal
    kernel = np.ones((3, 3), np.uint8)
    opening = cv2.morphologyEx(thresh, cv2.MORPH_OPEN, kernel, iterations=2)

    # Sure background
    sure_bg = cv2.dilate(opening, kernel, iterations=3)

    # Sure foreground
    dist_transform = cv2.distanceTransform(opening, cv2.DIST_L2, 5)
    _, sure_fg = cv2.threshold(dist_transform, 0.7 * dist_transform.max(), 255, 0)

    # Unknown region
    sure_fg = np.uint8(sure_fg)
    unknown = cv2.subtract(sure_bg, sure_fg)

    # Marker labelling
    _, markers = cv2.connectedComponents(sure_fg)
    markers = markers + 1
    markers[unknown == 255] = 0

    # Apply watershed
    markers = cv2.watershed(image, markers)
    segmented_mask = (markers > 1).astype(np.uint8)  # Convert to binary mask
    
    return segmented_mask

# Function to generate Grad-CAM heatmap on segmented knee
def apply_gradcam(image_path, mask):
    image = cv2.imread(image_path)
    image = cv2.resize(image, (224, 224))
    image = np.expand_dims(image, axis=0) / 255.0  # Normalize

    # Get last convolutional layer (Check your model's layer names)
    last_conv_layer = classifier_model.get_layer("conv2d_93")  # Change this layer name!

    grad_model = tf.keras.models.Model(
        [classifier_model.input], [last_conv_layer.output, classifier_model.output]
    )

    with tf.GradientTape() as tape:
        conv_output, predictions = grad_model(image)
        class_idx = np.argmax(predictions)
        class_channel = predictions[:, class_idx]

    grads = tape.gradient(class_channel, conv_output)

    # Generate heatmap
    heatmap = np.mean(grads, axis=-1)[0]
    heatmap = np.maximum(heatmap, 0)  # Remove negative values
    heatmap /= np.max(heatmap)  # Normalize

    # Resize heatmap to match mask size
    heatmap = cv2.resize(heatmap, (224, 224))

    # Convert mask to float before multiplication
    mask = mask.astype(np.float32)

    # Apply segmentation mask
    heatmap = heatmap * mask

    return heatmap

# Load and segment knee X-ray
image_path = "test1.png"
segmented_mask = segment_knee_watershed(image_path)

# Display segmentation result
plt.imshow(segmented_mask, cmap="gray")
plt.title("Segmented Knee Joint")
plt.show()

# Generate and display Grad-CAM heatmap
heatmap = apply_gradcam(image_path, segmented_mask)
plt.imshow(heatmap, cmap="jet")
plt.title("Heatmap on Segmented Knee Joint")
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.models import load_model

# Load trained classification model
classifier_model = load_model("my_model.h5")

def segment_knee_watershed(image_path):
    # Read image
    image = cv2.imread(image_path)
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Apply thresholding
    _, thresh = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

    # Noise removal
    kernel = np.ones((3, 3), np.uint8)
    opening = cv2.morphologyEx(thresh, cv2.MORPH_OPEN, kernel, iterations=2)

    # Sure background
    sure_bg = cv2.dilate(opening, kernel, iterations=3)

    # Sure foreground
    dist_transform = cv2.distanceTransform(opening, cv2.DIST_L2, 5)
    _, sure_fg = cv2.threshold(dist_transform, 0.7 * dist_transform.max(), 255, 0)

    # Unknown region
    sure_fg = np.uint8(sure_fg)
    unknown = cv2.subtract(sure_bg, sure_fg)

    # Marker labelling
    _, markers = cv2.connectedComponents(sure_fg)
    markers = markers + 1
    markers[unknown == 255] = 0

    # Apply watershed
    markers = cv2.watershed(image, markers)
    segmented_mask = (markers > 1).astype(np.uint8)  # Convert to binary mask

    return segmented_mask

def apply_gradcam(image_path, mask):
    image = cv2.imread(image_path)
    image = cv2.resize(image, (224, 224))
    image = np.expand_dims(image, axis=0) / 255.0  # Normalize

    # Get last convolutional layer (Check your model's layer names)
    last_conv_layer = classifier_model.get_layer("conv2d_93")  # Change this layer name!

    grad_model = tf.keras.models.Model(
        [classifier_model.input], [last_conv_layer.output, classifier_model.output]
    )

    with tf.GradientTape() as tape:
        conv_output, predictions = grad_model(image)
        class_idx = np.argmax(predictions)
        class_channel = predictions[:, class_idx]

    grads = tape.gradient(class_channel, conv_output)

    # Generate heatmap
    heatmap = np.mean(grads, axis=-1)[0]
    heatmap = np.maximum(heatmap, 0)  # Remove negative values
    heatmap /= np.max(heatmap)  # Normalize

    # Resize heatmap to match mask size
    heatmap = cv2.resize(heatmap, (224, 224))

    # Convert mask to float before multiplication
    mask = mask.astype(np.float32)

    # Apply segmentation mask
    heatmap = heatmap * mask

    return heatmap

def overlay_heatmap_on_image(image_path, heatmap):
    # Load original image
    original_image = cv2.imread(image_path)
    original_image = cv2.resize(original_image, (224, 224))

    # Convert heatmap to color map
    heatmap = np.uint8(255 * heatmap)  # Normalize to 0-255
    heatmap_colored = cv2.applyColorMap(heatmap, cv2.COLORMAP_JET)

    # Blend heatmap with original image
    overlay = cv2.addWeighted(original_image, 0.6, heatmap_colored, 0.4, 0)

    return overlay

# Load and segment knee X-ray
image_path = "test1.png"
segmented_mask = segment_knee_watershed(image_path)

# Generate Grad-CAM heatmap
heatmap = apply_gradcam(image_path, segmented_mask)

# Overlay heatmap on original image
overlay_image = overlay_heatmap_on_image(image_path, heatmap)

# Display the final image
plt.imshow(cv2.cvtColor(overlay_image, cv2.COLOR_BGR2RGB))
plt.title("Overlayed Heatmap on Knee X-ray")
plt.axis("off")
plt.show()
```
